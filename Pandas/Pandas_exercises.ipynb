```python
import numpy as np
import pandas as pd


#Para começar, precisamos importar a biblioteca pandas e carregar o dataset.
df = pd.read_csv('salaries.csv', sep=',')
#Após carregar o dataset, é útil visualizar as primeiras linhas para entender a estrutura dos dados.
print(df.head(2))
#Verifique os nulos e tipos de dados
df.isna().sum()
df.info()
#descubra o número de dados que o dataframe possui
df.count()
#qual é o basepay médio
df['BasePay']
df[df['BasePay'] == 'Not Provided']
df = df[~(df['BasePay'] == 'Not Provided')]
df['BasePay'] = df['BasePay'].astype(float)
df.info()
#Qual é o maior overtimepay do dataset
valor_maximo = df['OvertimePay'].astype(float).max()
df[df['OvertimePay']==valor_maximo]
#qual é o cargo do JOSEPH DRISCOLL
df[df['EmployeeName'] == 'JOSEPH DRISCOLL']
df[df['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
#quanto ele ganha incluindo benefícos?
df[df['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPay']
#pessoa que tem o maior salário base
df['BasePay'].max()
df[df['BasePay'] == df['BasePay'].max()]
#pessoa mais mal paga
df['BasePay'].min()
df[df['BasePay'] == df['BasePay'].min()]
#qual a média salarial por ano?
df['Year'].value_counts()
df.groupby('Year').agg({'BasePay': ['mean']})
#quantos títulos de trabalhos existem?
df['JobTitle'].nunique()
#quais são os 5 principais empregos mais comuns
df['JobTitle'].value_counts()
#quantas pessoas tem a palavra chefe em seu cargo?
def chefe(valor):
    if 'chief' in valor.lower():
        return True
    else:
        return False

df[df['JobTitle'].apply(chefe)]
```
